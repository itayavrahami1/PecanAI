```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
test_df = pd.read_csv('data/original_data/test.csv')
train_df = pd.read_csv('data/original_data/train.csv')
val_df = pd.read_csv('data/original_data/val.csv')

full_train_df = pd.concat([train_df, val_df])
```

Now we have 2 data sets: full_train_df, test_df. Letws explore the data. (shape, NaN's, statistics)

```python
print('HEAD\n------\n')

print('Train\n------\n',full_train_df.head())
print('Test\n------\n',test_df.head())
```

```python
print('INFO\n------\n')

print('Train\n------\n',full_train_df.info())
print('Test\n------\n',test_df.info())
```

```python
print('DESCRIPTION\n------\n')

print('Train\n------\n',full_train_df.describe().transpose())
print('Test\n------\n',test_df.describe().transpose())
```
